```python
def sentinel_read(sentinel_timestamp):
    sentinel_dir = os.path.join(root_dir, sentinel_timestamp)
    bands = glob.glob(os.path.join(sentinel_dir, '**', '*.jp2'), recursive=True)

    band_arrays = {}
    band_sources = {}

    for band_file in bands:
        fname = os.path.basename(band_file)

        if fname.endswith('TCI.jp2'):
            continue

        band_name = fname.split('_')[-1].replace('.jp2', '')

        src = rasterio.open(band_file)
        arr = src.read(1)

        # initialize lists
        if band_name not in band_arrays:
            band_arrays[band_name] = []
            band_sources[band_name] = []

        band_arrays[band_name].append(arr)
        band_sources[band_name].append(src)

    return sentinel_dir, band_arrays, band_sources
```

```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from rasterio.enums import Resampling

def visualize_index(band_arrays, band_sources, index_type="NDVI", downsample=5, cmap=None, band_labels=None):
    """
    Visualize satellite indices across multiple timestamps horizontally.
    Automatically resamples lower-resolution bands to match 10m NIR (B08).

    Parameters:
        band_arrays: dict of bands, e.g. {'B02': [...], 'B03': [...], ...}
        band_sources: dict of rasterio DatasetReaders for resampling
        index_type: "NDVI", "NDWI", "NDBI", "SAVI"
        downsample: int, pixel step to reduce size for memory safety
        cmap: matplotlib colormap (optional)
        band_labels: optional list of custom titles per timestamp
    """
    
    n = len(band_arrays['B04'])  # number of timestamps
    
    fig, axes = plt.subplots(1, n, figsize=(5*n, 5))
    if n == 1:
        axes = [axes]
    
    for idx, ax in enumerate(axes):
        # 10 m bands
        red = band_arrays['B04'][idx].astype(float)
        green = band_arrays['B03'][idx].astype(float)
        nir = band_arrays['B08'][idx].astype(float)
        
        # resample B11 and B12 (20 m) or any other mismatched band
        if 'B11' in band_arrays:
            src_swir = band_sources['B11'][idx]
            swir = src_swir.read(
                1,
                out_shape=(nir.shape[0], nir.shape[1]),
                resampling=Resampling.bilinear
            ).astype(float)
        else:
            swir = None
        
        # downsample all bands for memory
        red = red[::downsample, ::downsample]
        green = green[::downsample, ::downsample]
        nir = nir[::downsample, ::downsample]
        if swir is not None:
            swir = swir[::downsample, ::downsample]
        
        # calculate index
        itype = index_type.upper()
        if itype == "NDVI":
            img = (nir - red) / (nir + red + 1e-10)
            cmap = cmap or "RdYlGn"
        elif itype == "NDWI":
            img = (green - nir) / (green + nir + 1e-10)
            cmap = cmap or "Blues"
        elif itype == "NDBI":
            if swir is None:
                raise ValueError("SWIR band not available for NDBI")
            img = (swir - nir) / (swir + nir + 1e-10)
            cmap = cmap or "gray"
        elif itype == "SAVI":
            L = 0.5
            img = ((nir - red) * (1 + L)) / (nir + red + L + 1e-10)
            cmap = cmap or "RdYlGn"
        else:
            raise ValueError(f"Unknown index_type: {index_type}")
        
        # plot
        im = ax.imshow(img, cmap=cmap)
        title = f"T{idx}" if band_labels is None else band_labels[idx]
        ax.set_title(title)
        ax.axis('off')
    
    fig.colorbar(im, ax=axes, fraction=0.03, pad=0.04, label=itype)
    plt.tight_layout()
    plt.show()


# NDVI
visualize_index(band_arrays, band_sources, index_type="NDVI", downsample=5)

# NDWI
visualize_index(band_arrays, band_sources, index_type="NDWI", downsample=5)

# NDBI (works now!)
visualize_index(band_arrays, band_sources, index_type="NDBI", downsample=5)

# SAVI
visualize_index(band_arrays, band_sources, index_type="SAVI", downsample=5)

# Optional: label timestamps
dates = ["2017-08-04", "2017-08-14"]
visualize_index(band_arrays, band_sources, index_type="NDVI", downsample=5, band_labels=dates)
```
