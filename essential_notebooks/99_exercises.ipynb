   * *Hint:* 1 Hectare = 10,000 sqm. 1 Ton = 1,000 kg.
3. Print the result.

```python
# Write your solution here
plot_area_sqm = 5000
plot_area_hec = (1/10000)*plot_area_sqm
production_kg = 2500
production_tons = (1/1000)*production_kg

yield_tons_per_ha = production_tons/plot_area_hec

print(yield_tons_per_ha)
```

## 03
**Intermediate**

1. Using the formula in Exercise 02, create a function that takes `plot_area_sqm` and `production_kg` as parameters.
2. Return a dictionary as output with the following keys:
    - `plot_area_sqm`
    - `production_kg`
    - `yield_tons_per_ha`

```python
# Write your solution here

def _yield_tons_per_ha(plot_area_sqm, production_kg):
    plot_area_hec = (1/10000)*plot_area_sqm
    production_tons = (1/1000)*production_kg

    yield_tons_per_ha = production_tons/plot_area_hec

    return {
        "plot_area_sqm": plot_area_sqm,
        "production_kg": production_kg,
        "yield_tons_per_ha": yield_tons_per_ha
    }

_yield_tons_per_ha(5000, 2500)
```

```python
# Write your solution here
from datetime import datetime

date_str = "May 1, 2026"

future_date = datetime.strptime(date_str, "%B %d, %Y")

now = datetime.now()

print("now", now)

diff = future_date - now

print(diff.days)
```
